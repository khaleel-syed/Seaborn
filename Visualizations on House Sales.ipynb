```python
import numpy as np
import pandas as pd
```

```python
house = pd.read_excel("housesales.xlsx")
house.head(5)
```

```python
house.info()
```

## Numerical Variable

- univariate
- Bi-variate
- Multi -variate

### Univariate Distribution

```python
import seaborn as sns
```

```python
sns.distplot(house['sqft'])
```

```python
# Distribution of a list price
sns.distplot(house['listprice'])
```

```python
sns.distplot(house['listprice'],kde=False)
```

## Bivariate Distribution

```python
sns.jointplot(x=house['sqft'],y=house['listprice'])
```

```python
sns.jointplot(x=house['lotsize'],y=house['listprice'])
```

```python
sns.jointplot(x=house['daysonmarket'],y=house['listprice'])
```

```python
sns.jointplot(x=house['sqft'],y=house['listprice'],kind='kde')
```

## Multi variate Distribution

```python
data = house[['sqft','beds','hometype','daysonmarket','listprice']]
sns.pairplot(data=data,hue='hometype')
```

## Categorical Variable

#### categorical Scatter plot
- strip
- swarm
#### Categorical Distribution
- Boxplot
- Violine
#### Statistical Estimation
- barplot
- countplot

```python
sns.stripplot(y=house['listprice'])
```

```python
sns.stripplot(x=house['hometype'],y=house['listprice'])
```

```python
sns.stripplot(x=house['hometype'],y=house['listprice'],hue=house['parkingtype'])
```

```python
sns.swarmplot(x='hometype',
             y='listprice',
             data=house)
```

```python
sns.swarmplot(x='hometype',
             y='listprice',
             data=house,
             hue='parkingtype')
```

```python
sns.boxplot(x='listprice',data=house)
```

```python
sns.boxplot(x='listprice',y='hometype',data=house)
```

```python
sns.boxplot(x='zip',y='listprice',data=house)
```

```python
house['listprice'].quantile([0.25,0.50,0.75])
```

```python
sns.violinplot(x='listprice',data=house)
```

```python
sns.barplot(x='hometype',y='listprice',data=house)
```

```python
sns.barplot(x='hometype',y='listprice',hue='parkingtype',data=house)
```

```python
sns.countplot(x='hometype',data=house)
```

```python
sns.countplot(x='hometype',hue='parkingtype',data=house)
```

```python
sns.lineplot(x='daysonmarket',y='listprice',data=house)
```
